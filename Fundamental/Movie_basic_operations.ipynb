### Row Filtering

```python
year_range = df.selectExpr("min(release_year) as min_year", "max(release_year) as max_year")
display(year_range)
```

```python
min_max_year = df.selectExpr("min(release_year) as min_year", "max(release_year) as max_year").collect()
min_year = min_max_year[0]['min_year']
max_year = min_max_year[0]['max_year']

earliest_latest_revenue = df.filter(df.release_year.isin([min_year, max_year])) \
    .select("title", "release_year", "revenue") \
    .orderBy("release_year")
display(earliest_latest_revenue)
```

```python
from pyspark.sql.functions import col

# Filter rows where release_year is between 2000 and 2010 (inclusive)
df_filtered = df.filter(
    (col("release_year") >= 2000) & (col("release_year") <= 2010)
)

df_filtered.show()
```

```python
df.filter(col("release_year").between(2000, 2010)).show()
```
